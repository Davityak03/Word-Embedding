```python
#importing one_hot from tensorflow
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=['the glass of milk',
     'the glass of juice',
     'the cup of tea',
     'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
voc_size=10000#size of the vocabulary
```

```python
#one hot representaion
onehot_repr=[one_hot(words,voc_size)for words in sent]
print(onehot_repr)
```

**word embedding representation**

```python
#importing the neccessary libraries
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
import pandas as pd
```

```python
sent_length=8 #sentence length after padding
embedded_docs=pad_sequences(onehot_repr,padding="pre",maxlen=sent_length) #padding=pre means zero will be infront
print(embedded_docs)# sentences after padding
#padding is done to make the sentences length equal to one another
```

```python
table1=pd.DataFrame(embedded_docs)
table1
```

```python
dim=15#dimensions of the each vector that will represent the word
```

```python
#creating the embedding model
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile("adam","mse")
```

```python
model.summary()
```

```python
#word embedding table
print(model.predict(embedded_docs))
```
